Основные значения:

* CRIM: This is the per capita crime rate by town

* AGE: This is the proportion of owner-occupied units built prior to 1940
* DIS: This is the weighted distances to five Boston employment centers
* RAD: This is the index of accessibility to radial highways
* TAX: This is the full-value property-tax rate per 10,000 dollars
* PTRATIO: This is the pupil-teacher ratio by town
* B: This is calculated as 1000(Bk — 0.63)², where Bk is the proportion of people of African American descent by town
* LSTAT: This is the percentage lower status of the population

Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
# линейная регрессия 
# предсказываемая величина (y) - MEDV, независимая величина (x) LSTAT
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
medv = np.array(data['MEDV']).reshape(-1,1)
lstat = np.array(data['LSTAT']).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
# x = lstat
# y = medv
x_train, x_test, y_train, y_test = train_test_split(
    lstat,
    medv,
    test_size = 0.2
)
```

```python
model.fit(x_train, y_train)
```

```python
predicted_y = model.predict(x_test)
```

```python
model.score(x_test, y_test)
# модель дает точность 54,6% на тестовых данных
```

```python
# точечный график данных и полученной прямой линейной регрессии
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title('Линейная регрессия от одной переменной')

x = np.linspace(-50, 200, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(x_train, y_train, 'r.')
plt.plot(x_test, y_test, 'go')
plt.xlim(0,50)
plt.ylim(0,50)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
toss_cnt = 10
list_random = np.random.randint(1,7, toss_cnt)
n, bins, patches  = plt.hist(list_random, 10)
# равномерное распределение
```

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
number_of_sums = 1000
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    X1[i] = np.sum(np.random.randint(1,7, number_of_sums))
plt.hist(X1, 50)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.show()  
# Суммы элементов элементов выборок стремится к нормальному распределению!

# График изменился согласно закону о больших числах, так как сумма достаточно большого количества 
# слабо зависимых случайных величин, имеющих примерно одинаковые масштабы, имеет распределение, 
# близкое к нормальному.
# А величины при подбрасывании кубика имеют примерно одинаковые масштабы и слабозависимы друг от друга
```
